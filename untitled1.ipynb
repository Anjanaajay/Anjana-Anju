```python
def print_positive_numbers(input_list):
    """Prints all positive numbers in the given list."""
    positive_numbers = [num for num in input_list if num > 0]
    print(*positive_numbers, sep=', ')

# Example usage matching the input/output format
list1 = [12, -7, 5, 64, -14]
print("Input: list1 =", list1, "Output:", end=' ')
print_positive_numbers(list1)

list2 = [12, 14, -95, 3]
print("Input: list2 =", list2, "Output:", end=' ')
print_positive_numbers(list2)
```
